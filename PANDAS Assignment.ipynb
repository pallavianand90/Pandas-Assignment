```python
# Find out how many male and female passenger was onboarded
print("male:",len(df[df["Sex"]=="male"]))
print("female:",len(df[df["Sex"]=="female"]))
```

```python
# How many survived we have
print("Survived:",len(df[df['Survived']==1]))
```

```python
# how many casuality we have
print("Casuality:",len(df[df['Survived']==0]))
```

```python
# What is name of a person who is the eldest one.
print(df[df['Age']==max(df["Age"])]["Name"])
```

```python
# how many passenger do we have in first , second and third class
print("First Class:",len(df[df["Pclass"]==1]))
print("Second Class:",len(df[df["Pclass"]==2]))
print("Third Class:",len(df[df["Pclass"]==3]))
```

```python
# how many person we have whose name starts with 's'.
print("Number of person whose name starts with S:",len(df[df["Name"].str[:1]=="S"]))
```

```python
# Try to create a new column which is a summation of "SibSp" and "parch"
df["Sibsp_parch"] = df["SibSp"] + df["Parch"]
df
```

```python
# How many person do we have below age of 25 .
print("Number of person below age of 25:",len(df[df["Age"]<25]))
```

```python
# Hpw many person died whose age was less then 40.
print("Number of person died whose age < 40:",len(df[(df.Age<40) & (df.Survived==0)]))
```

```python
# From a cabin column sperate text and numeric value.
df['Cabin']
```

```python
df['Section_Number'] = df['Cabin'].str.replace('([A-Z]+)', '')
df['Section_Letter'] = df['Cabin'].str.extract('([A-Z]+)')
df
```
